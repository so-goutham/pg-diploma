```python
import pandas as pd
data=pd.read_csv('/content/normalized_data.csv')
display(data)
```

```python
data.columns
```

```python
features=data.drop(columns='Ex_Showroom_Price')
target=data['Ex_Showroom_Price']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.2,random_state=45)
```

```python
 def lr(xt,xs,yt,ys):
  from sklearn.linear_model import LinearRegression
  li=LinearRegression()

  model=li.fit(xt,yt)
  ypred=li.predict(xs)
  from sklearn.metrics import  mean_squared_error,r2_score
  print(f"mse : {mean_squared_error(ypred,ys)}")
  print(f"r2_score: {r2_score(ypred,ys)}")
  return model
```

```python
lr(xtrain,xtest,ytrain,ytest)
```

```python
def knr(xt,xs,yt,ys):
  from sklearn.neighbors import KNeighborsRegressor
  kn=KNeighborsRegressor()

  model=kn.fit(xt,yt)
  ypred=kn.predict(xs)
  from sklearn.metrics import  mean_squared_error,r2_score
  print(f"mse : {mean_squared_error(ypred,ys)}")
  print(f"r2_score: {r2_score(ypred,ys)}")
  return model
```

```python
knr(xtrain,xtest,ytrain,ytest)
```

```python
def svr(xt,xs,yt,ys):
  from sklearn.svm import SVR
  sv=SVR()

  model=sv.fit(xt,yt)
  ypred=sv.predict(xs)
  from sklearn.metrics import  mean_squared_error,r2_score
  print(f"mse : {mean_squared_error(ypred,ys)}")
  print(f"r2_score: {r2_score(ypred,ys)}")
  return model
```

```python
svr(xtrain,xtest,ytrain,ytest)
```

```python
def dt(xt,xs,yt,ys):

  from sklearn.model_selection import RandomizedSearchCV
  from scipy.stats import randint
  from sklearn.tree import DecisionTreeRegressor
  from sklearn.metrics import r2_score,mean_squared_error

  dc=DecisionTreeRegressor()
  param={
  'max_depth': randint(3, 20),
  'min_samples_split': randint(2, 20),
  'min_samples_leaf': randint(1, 10),
  'max_features': ['sqrt', 'log2']
  }
  random_search = RandomizedSearchCV(estimator=dc,param_distributions=param,n_iter=20,cv=5,scoring='neg_mean_squared_error',random_state=42,n_jobs=-1)

  random_search.fit(xt, yt)
  best_params=random_search.best_params_
  bp=[]
  for i in best_params.values():
    bp.append(i)


  md,mf,msl,mss=bp


  model=DecisionTreeRegressor(max_depth=md,max_features=mf,min_samples_leaf=msl,min_samples_split=mss)
  model.fit(xt,yt)
  ypred=model.predict(xs)
  print(r2_score(ypred,ys))
  print(mean_squared_error(ypred,ys))

  return model
```

```python
dt(xtrain,xtest,ytrain,ytest)
```

```python
def rf(xt,xs,yt,ys):

  from sklearn.model_selection import RandomizedSearchCV
  from scipy.stats import randint
  from sklearn.ensemble import RandomForestRegressor
  from sklearn.metrics import r2_score,mean_squared_error

  rc=RandomForestRegressor()
  param={
  'n_estimators':randint(50,500),
  'max_depth': randint(3, 20),
  'min_samples_split': randint(2, 20),
  'min_samples_leaf': randint(1, 10),
  'max_features': ['sqrt', 'log2']
  }
  random_search = RandomizedSearchCV(estimator=rc,param_distributions=param,n_iter=20,cv=5,scoring='neg_mean_squared_error',random_state=42,n_jobs=-1)

  random_search.fit(xt, yt)
  best_params=random_search.best_params_
  bp=[]
  for i in best_params.values():
    bp.append(i)


  md,mf,msl,mss,ne=bp


  model=RandomForestRegressor(n_estimators=ne,max_depth=md,max_features=mf,min_samples_leaf=msl,min_samples_split=mss)
  model.fit(xt,yt)
  ypred=model.predict(xs)
  print(r2_score(ypred,ys))
  print(mean_squared_error(ypred,ys))
  return model
```

```python
rf(xtrain,xtest,ytrain,ytest)
```
