```python
import pandas as pd
data=pd.read_csv('/content/Bike_prices_after_outlier_removal.csv')
display(data)
```

```python
value_count=['Brand','Model','Year','Seller_Type','Owner']
```

```python
for v in value_count:
  dat=data.value_counts(v)
  display(dat)
```

```python
order=data['Year'].unique()
order.sort()
print(order)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord=OrdinalEncoder(categories=[order])
data['Year']=ord.fit_transform(data[['Year']])

display(data)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Brand'] = le.fit_transform(data['Brand'])
data['Model'] = le.fit_transform(data['Model'])
display(data)
```

```python
lis_on=['Seller_Type','Owner']
```

```python
from sklearn.preprocessing import OneHotEncoder

for o in lis_on:
    on=OneHotEncoder(sparse_output=False,drop='first')

    st=on.fit_transform(data[[o]])

    st=pd.DataFrame(st,columns=on.get_feature_names_out([o]))

    data=pd.concat((data,st),axis=1)
    data=data.drop(columns=o)
data
```

```python
data.to_csv('encoded_bike_prices.csv',index=False)
```
