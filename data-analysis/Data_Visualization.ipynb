## Matplotlib

Matplotlib is an easy-to-use, low-level data visualization library that is built on NumPy arrays. It provides a lot of flexibility.

    Scatter Plot  - plt.scatter()
    Line Chart    - plt.plot()
    Bar Chart	  - plt.bar()
    Histogram	  - plt.hist()
    Boxplot     -plt.boxplot()
    Pie Chart   - plt.pie()

```python
# Scatter plot with day against tip
plt.scatter(df['total_bill'], df['tip'])

# Adding Title to the Plot
plt.title("Total bill - tip plotting")

# Setting the X and Y labels
plt.xlabel('total_bill')
plt.ylabel('Tip')

plt.show()
```

```python

# Line plot with day against tip
plt.plot(df['size'],label="size")
plt.plot(df['total_bill'],label="total_bill")
plt.plot(df['tip'],label="Tip")

# Adding Title to the Plot
plt.title("Line Plot")

plt.legend()
plt.show()
```

this is lineplot . lineplot shows that data of size total bill and tip


```python
# Bar chart with day against tip
plt.bar(df['day'], df['tip'],label="day-tip")
plt.bar(df['time'], df['tip'],label='time-tip')
plt.bar(df['smoker'], df['tip'],label='smoker-tip')
plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.legend()
# Adding the legends
plt.show()
```

this one barchart. here data shows of day-tip,time-tip,smoker-tip barchart differentiate with colors

```python
# histogram of total_bills
plt.hist(df['total_bill'])

plt.title("Histogram")

# Adding the legends
plt.show()
```

## Seaborn

    Scatter Plot  - sns.scatterplot()
    (Easy use of the hue argument for coloring by category (e.g., sex))

    Line Plot - sns.lineplot()
    (Simplifies creating line charts, even without specifying X and Y)

    Bar Plot  - sns.barplot()
    (Easily incorporates categorical variables for comparison)

    Histogram - sns.histplot()
    (Includes an option for a Kernel Density Estimate (kde=True))


```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df=sns.load_dataset('tips')
display(df)

```

```python
# scatter plot

#plt.scatter(df['day'],df['tip'])#matplot
sns.scatterplot(x='tip', y='total_bill', data=df)
# setting the title using Matplotlib
plt.title('Tip - Total Bill')
plt.show()
```

this plot was data of tip and total bill. here bill count was higher an d tip was low. and here the scatter plot shows outliers

```python
sns.regplot(x='tip', y='total_bill', data=df)
plt.title("Regression Line")
```

this one rigplot. this plot exactly looks like scaterplot the differnce is the line regression line

```python
# draw lineplot



sns.lineplot(x="tip", y="day", data=df)

# setting the title using Matplotlib
plt.title('Line Plot')

plt.show()
```

here is  lineplot. the line plot shows data of tip and day  

```python
sns.boxplot(df['tip'])
plt.show()
```

here ploting was boxplot. the boxplot shows the data of tip. boxplot shows upperside 25% and lower side shows 25%. middle blue box shows
50% data. then outside dots that ouliers


```python
plt.bar(df['day'],df['tip'])
plt.show()
```

bargraph. bargraph shows the data of sunday how much, stuerday how much,
thursday how much , friday how much

```python
# barplot
sns.barplot(x='day',y='total_bill', data=df, hue='smoker')

plt.show()
```

here barplot. it shows data of this days thhursday, friday, saturday
sunday, how many smokers on that day

```python
# histplot

sns.histplot(x='tip', data=df, kde=True,hue='sex') # kde- Kernal Density Estimation

plt.show()


```

histogram. here data shows sex male and female count and they give the  tip %.

skew - distribution

    Right skewed - data-> left - mean>median>mode
    left skewed - data-> right - mode>median>mean

Kurtosis - peakedness

    leptokurtic - high
    mesokurtic - normal
    platykurtic - flat

```python

plt.figure(figsize=(8, 5))
sns.kdeplot(df, fill=True, color="skyblue", linewidth=2)

# Add labels and title
plt.title("Kernel Density Estimate (KDE) Plot", fontsize=14)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Show the plot
plt.show()
```
