## While Loop in Python

1. **Print numbers 1 to 10**
   Write a `while` loop to print all numbers from **1 to 10**.


````python
number=1
while number <=10:
  print(number)
  number +=1
````

2. **Countdown timer**
   Use a `while` loop to print numbers from **10 down to 1**.


````python
count=10
while count >=1:
  print(count)
  count -=1
````

6. **Reverse a number**
   Use a `while` loop to reverse the digits of a given number.
   *(Example: Input → 1234, Output → 4321)*


````python
n=1234
while n!=0:
  r=n%10
  n=n//10
  print(r,end="")
````

9. **Sum of digits**
   Use a `while` loop to calculate the sum of digits of a given number.
   *(Example: 123 → 1+2+3 = 6)*

````python
n=int(input("enter the number"))
sum=0
while n!=0:
  r=n%10
  n=n//10
  sum=r+sum
print(sum)
````
